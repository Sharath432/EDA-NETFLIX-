3. DEALING WITH MISSING DATA

```python
#droping the null columns -- director

no_director = df[df['director'].isna()].index
df.drop(no_director, axis = 0, inplace = True)
df
```

```python
# dropping director columns
df = df.dropna(subset = 'director', axis = 0)
df
```

```python
df['cast'].fillna('no_cast', inplace = True)
df.isna().sum()
```

```python
df['country'] = df['country'].fillna(df['country'].mode())
df['country']
```

```python
df.head()
```

```python
#Extracting data from duration column with split and extract method.

df_movie = df[df['type'] == 'Movie']
df_movie['duration'].str.split(expand = True)[0].astype('int')
df_movie['country'].fillna(method = 'ffill', inplace = True)
df_movie.isna().sum()

df_movie['rating'].fillna(method = 'bfill', inplace = True)
df_movie.isna().sum()
```

```python
#Extracting data from duration column with split and extract method.

df_movie = df[df['type'] == 'Movie']
df_movie['duration_minutes'] = df_movie['duration'].str.split(expand = True)[0].astype('int')
df_movie['duration_minutes']

df_movie.drop('duration', axis = 1, inplace = True)
df_movie.head()
```
